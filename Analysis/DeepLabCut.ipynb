```python
#GUIs don't work on the cloud, so we will supress wxPython: 
import os
os.environ["DLClight"]="True"

import deeplabcut
```

```python
path_config_file = '/content/realHand/config.yaml'
```

#### Between these two sections, the file system should be updated

# Create Dataset network

```python
deeplabcut.create_multianimaltraining_dataset(
    path_config_file,
    num_shuffles=1,
    net_type="dlcrnet_ms5"
)
```

# Train network

```python
deeplabcut.train_network(
    path_config_file,
    displayiters=10,
    saveiters=50,
    maxiters=5000,
    allow_growth=True,
)
```

# Evaluate the network

```python
%matplotlib notebook
deeplabcut.evaluate_network(path_config_file,plotting=True)
```

# Analyze the videos

```python
videofile_path = ['/content/realHand/videos/WIN_20211102_16_19_47_Pro.mp4'] #Enter the list of videos to analyze.
deeplabcut.analyze_videos(path_config_file,videofile_path, videotype='.mp4')
```

# Label the video

```python
deeplabcut.create_labeled_video(path_config_file,videofile_path)
```

# Create trackers

```python
deeplabcut.convert_detections2tracklets(
    path_config_file,
    videofile_path,
    track_method="ellipse",
)
```

# Stitch trackers

```python
deeplabcut.stitch_tracklets(
    path_config_file,
    videofile_path,
    track_method="ellipse",
    min_length=5,
)
```

# Create labeled video

```python
deeplabcut.create_labeled_video(
    path_config_file,
    videofile_path,
    color_by="individual",
    keypoints_only=False,
    trailpoints=0,
    draw_skeleton=True,
    track_method="ellipse",
)
```
